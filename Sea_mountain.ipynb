```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/mountains_vs_beaches_preferences.csv')
df
```

```python
df.drop(['Gender','Education_Level','Pets'],axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Preferred_Activities']=lab.fit_transform(df['Preferred_Activities'])
df['Location']=lab.fit_transform(df['Location'])
df['Favorite_Season']=lab.fit_transform(df['Favorite_Season'])
```

```python
df.dtypes
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
x_test
```

```python
y_test
```

```python
y_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
knn=KNeighborsClassifier(n_neighbors=7)
bays=GaussianNB()
sv=SVC()
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
lst=[knn,bays,sv]
```

```python
for i in lst:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print("Model is",i)
  print(y_pred)
  print("*"*100)
  print("Confusion matrix is...........")
  print(confusion_matrix(y_test,y_pred))
  print("Accuracy score is.............")
  print(accuracy_score(y_test,y_pred))
  print("classification matrix is...........")
  print(classification_report(y_test,y_pred))
```
